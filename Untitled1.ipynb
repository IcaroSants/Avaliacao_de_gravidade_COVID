```python
import os
import pandas as pd
```

```python
diretorio_atual = os.getcwd()
pasta_img = os.path.join(diretorio_atual,'Dataset')
lista_img = os.listdir(pasta_img)
```

```python
pac_nomes = []
for arq in lista_img:
    pac_nomes.append(arq[:-4])
```

```python
pac_nomes
```

```python
lista_caminhos  = []
for arq in lista_img:
    path = os.path.join(pasta_img,arq)
    lista_caminhos.append(path)
```

```python
lista_caminhos
```

```python
mapeamento = {paciente:caminho for paciente, caminho in zip(pac_nomes,lista_caminhos)}
```

```python
mapeamento
```

```python
df = pd.read_csv('metadados.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Path'] = df['Patient'].map(mapeamento)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.to_csv('metadados_Dataset')
```
